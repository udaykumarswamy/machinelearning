```python
#as we see the data , it has been in the different scaling so we need to scale them in a single format scale
#this has to be done as SVM tries to calculate the max distace of the clustered data points
#this scalling has to be only after spliting the data into features and target variables and train and test
#because then the info from test will leak to train model
```

```python
#now lets split the data into feature and target
feature = voiceDataFrame.drop(columns=['label'],axis=1)
taget = voiceDataFrame['label']
```

```python
feature_train,feature_test,target_train,target_test = train_test_split(feature,taget,test_size=0.3,random_state=42) 
```

```python
#scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
feature_train_scaled=scaler.fit_transform(feature_train)
#here we need to do only transform as test data may be revealed
#we can also use minMaxScalers
feature_test_scaled = scaler.transform(feature_test)
```

```python
#for svm import sklearn.svm module 
from sklearn.svm import SVC
```

```python
#the model can be optimised the more accuracte by changing the 'c' and gamma
# 'c' and gamma if we increase them it will increase the train model accuracy will increase 
#however the model tend to overfitted
svc = SVC()

svc.fit(feature_train_scaled,target_train)
target_pred_train = svc.predict(feature_train_scaled)
target_pred_test = svc.predict(feature_test_scaled)
```

```python
#now lets evealuate the model accuracy
#svc.score is another to get the accuracy metrics
accuracy_train = svc.score(feature_train_scaled,target_train).round(2)
accuracy_test = svc.score(feature_test_scaled,target_test).round(2)
accuracy_train,accuracy_test
```
